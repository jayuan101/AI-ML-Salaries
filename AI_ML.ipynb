```python
X = df[predictor]
```

```python
X
```

```python
y = df["salary_in_usd"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
salaries_lr = LinearRegression()
```

```python
salaries_lr.fit(X_train, y_train) #training the model
```

```python
train_score = salaries_lr.score(X_train, y_train)
print('Linear Regression on trained data = ', train_score)
```

```python
test_score = salaries_lr.score(X_test, y_test)
print('Linear Regression on test data = ', test_score)
```

```python
salaries_lr.coef_
```

```python
salaries_lr.intercept_
```

```python
y_predict_train = salaries_lr.predict(X_train)
```
